### Choose a Clustering Algorithm

K means clustering partitions the data into a prespecified number of clusters by minimizing the wintin-cluster sum of squares. 

```python
## Your Answer Here

np.random.seed(10)

wcss = []

# Test different numbers of clusters
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(nhanes_scaled)
    wcss.append(kmeans.inertia_)  # WCSS (inertia) for the current k

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.grid(alpha=0.5)
plt.show()

```

### Cluster nhanes

I used the elbow method to determine the number of clusters. There are diminishing returns beyond 4 clusters, so this is what I chose. 

```python


kmeans = KMeans(n_clusters=4,
                n_init=10,
                max_iter=300 
                ).fit(nhanes_scaled)
```

### Plot your clusters on top of the BMI v. Income Poverty Ratio Plot

```python
## Your Answer Here

nhanes_scaled['Cluster'] = kmeans.fit_predict(nhanes_scaled[['INDFMPIR', 'BMXBMI']])

# Create the scatter plot with clusters as the hue
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    x="INDFMPIR",
    y="BMXBMI",
    hue="Cluster",
    palette="tab10",  
    data=nhanes_scaled,
    alpha=0.8
)

# Customize plot labels and title
ax.set(
    xlabel="Income to Poverty Line Ratio",
    ylabel="Body Mass Index",
    title="BMI v. Income-Poverty Ratio with KMeans Clusters"
)
plt.legend(title="Cluster", loc="best")
plt.grid(alpha=0.5)
plt.show()
```

In general, the clusters are defined by the four quadrants of the plot, such that high BMI/high income to poverty, high BMI/low income to poverty, low BMI/high income to poverty and low BMI/low income to poverty are clustered together. There seems to be some binning happening at the high end of income to poverty which may skew these slightly.

### Retrain the clustering algorithm on PCA components and plot clusters on your 2D scatter

```python
## Your Answer Here

#create dataframe with PCA components
pca_df = pd.DataFrame(data = PCA_five_fit,
                     columns = ['component 1',
                                        'component 2',
                                        'component 3',
                                        'component 4',
                                        'component 5'])


kmeans_pca = KMeans(n_clusters=4,
                n_init=10,
                max_iter=300 
                ).fit(pca_df)

kmeans_clusters = kmeans_pca.fit_predict(pca_df)
# Scatterplot of the first two principal components with KMeans clusters
plt.figure(figsize=(8, 6))
for cluster in np.unique(kmeans_clusters):
    plt.scatter(
        PCA_five_fit[kmeans_clusters == cluster, 0],
        PCA_five_fit[kmeans_clusters == cluster, 1],
        label=f'Cluster {cluster}', 
        alpha=0.7, 
        s=50
    )

# Add cluster centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], 
            s=200, c='black', marker='X', label='Centroids')

# Customize plot labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D Scatterplot of KMeans Clusters')
plt.legend(title='Cluster')
plt.grid(alpha=0.5)
plt.show()
```
